```python
# herança Multipla e herança Multinivel

# Classe Avô
class Animal():
  def __init__(self, nome):
    self.nome = nome

  def comer(self):
    print(f"{self.nome} está comendo!")

# Classes Pai
class Predador(Animal):
  def cacando(self):
    print(f"O animal {self.nome} que é predador está cacando!")

class Presa(Animal):
  def fugindo(self):
    print(f"O animal {self.nome} que é presa está fugindo!")

#Classes Filho
class Coelho(Presa):
  pass

class Tigre(Predador):
  pass

class Golfinho(Predador, Presa): #duas heranças
  pass

coelho1 = Coelho('Bunny')
tigre1 = Tigre('Teo')
golfinho1 = Golfinho('Billy')

coelho1.fugindo()
tigre1.cacando()
golfinho1.cacando()
golfinho1.fugindo()
```
