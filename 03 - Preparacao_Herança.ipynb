```python
class Pessoa:
  #construtor
  def __init__(self, nome, idade, cargo):
    self.nome = nome
    self.idade = idade
    self.cargo = cargo
  #funções
  def informacoes(self):
    print(f'Nome {self.nome}')
    print(f'Idade {self.idade}')
    print(f'Cargo {self.cargo}')

  def promover(self, novo_cargo):
    self.cargo = novo_cargo
    print(f'\nO colaborador {self.nome} foi promovido para {self.cargo}')

  def atualizar_dade(self, nova_idade):
    if nova_idade > self.idade:
      self.idade = nova_idade
      print(f'\nA idade do colaborador {self.nome} foi atualizada para {self.idade}')
    else:
      print('A nova idade deve ser diferente a idade anterior')


colaborador1 = Pessoa ('Ana', 25, 'Analista')
colaborador2 = Pessoa('Carlos', 45, 'Gerente')

# Criando instâncias fora da classe
colaborador2.informacoes()
colaborador2.promover('Diretor')
colaborador2.atualizar_dade(49)
```
