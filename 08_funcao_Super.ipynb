```python
#Sistema de Escola

class Escola():
  def __init__(self, nome, idade, status):
    self.nome = nome
    self.idade = idade
    self.status = status

  def Apresentar(self):
    print(f'Olá, meu nome é {self.nome} e tenho {self.idade} anos')

class Aluno(Escola):
  def __init__(self, nome, idade, matricula, status, ano):
    super().__init__(nome, idade, status)
    self.matricula = matricula
    self.ano = ano

  def Apresentar(self):
    super().Apresentar()
    print(f'Olá, meu nome é {self.nome}, tenho {self.idade} anos e meu ano é {self.ano}') #subscreveu o método do pai

class Professor(Escola):
  def __init__(self, nome, idade, matricula, status, disciplina):
    super().__init__(nome, idade, status)
    self.matricula = matricula
    self.disciplina = disciplina

class Assistente(Escola):
  def __init__(self, nome, idade, status, bloco):
    super().__init__(nome, idade, status)
    self.bloco = bloco

a1 = Aluno('Marcos', 13, 2586, 'Ativo', 8 )
p1 = Professor('Roberto', 34, 1234, 'Ativo', 'Matematica')
as1 = Assistente('Maria', 23, 'Ativo', 'Bloco A')

print(a1.nome)
print(a1.idade)
print(a1.matricula)
print(a1.ano)
print()
a1.Apresentar()

print('--------------------------')


print(p1.nome)
print(p1.idade)
print(p1.matricula)
print(p1.disciplina)
print()
p1.Apresentar()

print('--------------------------')


print(as1.nome)
print(as1.idade)
print(as1.bloco)
as1.Apresentar()
```
