```python
class Pessoa1:
  def microfone(self):
    print('Meu nome é Ana')

class Pessoa2:
  def microfone(self):
    print('Meu nome é Paulo')

sala = [Pessoa1(), Pessoa2()]

for pessoa in sala:
  pessoa.microfone()
```
