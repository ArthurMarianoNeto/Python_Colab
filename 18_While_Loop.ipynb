```python
'''
While loop
Excelente para loops dependentes de uma condição
Publicar um produto com uma comissão de 10% se for acima de $20,00
'''

valor = int(input('Digite o valor do seu produto: '))
#print(valor)
while valor > 20:
  valor = (valor * 0.10) + valor
  print(f'O valor final do seu produto é {valor}')
  break
```
