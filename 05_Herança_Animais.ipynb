```python
class Animal:
  def __init__ (self, nome, cor, especie): #parâmetros
    self.nome = nome
    self.cor = cor
    self.especie = especie

#método
  def apresentar(self):
    print(f'Eu sou o {self.especie} chamado(a) {self.nome}')

#classe filha
class Gato(Animal):
  def emitir_som(self):
    print('Miau')

  def arranhar(self):
    print('Estou arranhando')

class Cachorro(Animal):
  def emitir_som(self):
    print('Au Au')
  def abanar_rabo(self):
    print('Estou abanando o rabo')

class Galinha(Animal):
  def emitir_som(self):
    print('Cócóricó')

#Objeto ou Instancia
gato1 = Gato('Felix', 'Branco',  'Siamese')
gato1.apresentar() #este método está dentro da classe pai
gato1.emitir_som()
gato1.arranhar()

cachorro1 = Cachorro('Bob', 'Marrom', 'Labrador')
cachorro2 = Cachorro('Dylan', 'Preto', 'Poodle')
cachorro1.apresentar()
cachorro2.apresentar()
cachorro1.emitir_som()
cachorro1.abanar_rabo()

cachorro2.apresentar()
cachorro2.emitir_som()
cachorro2.abanar_rabo()

galinha1 = Galinha('MaryLu', 'Branca', 'Gaivota')
galinha1.apresentar()
galinha1.emitir_som()
```
