```python
class Casa:
    def __init__(self,cor, quartos, banheiros):
      self.cor = cor
      self.quartos = quartos
      self.banheiros = banheiros

    def mostrar_cor(self):
      print(f'A cor da casa {self.cor}')

    def mostrar_quartos(self):
      print(f'A casa tem {self.quartos} quartos')

    def mostrar_bancheiros(self):
      print(f'A casa tem {self.banheiros} banheiros')

    def adicionar_quartos(self):
      self.quartos += 1
      print(f'Esta casa tem {self.quartos} quartos')

    def pintar_casa(self, nova_cor):
      print(f'Pintando a casa de {self.cor} para {nova_cor}')


casa1 = Casa('Azul', 4, 3)
casa2 = Casa('Vermelha', 6, 9)

print('\ncasa1.cor:')
casa1.mostrar_cor()
casa1.mostrar_quartos()
casa1.mostrar_bancheiros()
casa1.adicionar_quartos()
casa1.pintar_casa('Verde')

print('\ncasa1.cor:')
casa2.mostrar_cor()
casa2.mostrar_quartos()
casa2.mostrar_bancheiros()
```

```python
class Carro:
 #-----------------------------------------------
 #construtores
  def __init__(self, cor, ano):
    self.cor = cor
    self.ano = ano
    self.ligado = True
    self.seta = None #campo vazio
#-----------------------------------------------
#funções
  def informacoes(self):
    print(f'A cor do carro é {self.cor}')

  def ligar(self):
    if not self.ligado:
      self.ligado = True
      print('Carro foi ligado agora')
    else:
      print('Carro já  estava ligado')

  def desligar(self):
    if self.ligado:
      self.ligado = False
      print('Carro foi desligado agora')
    else:
      print('Carro já estava desligado')

  def ligar_seta(self, direcao):
    if not self.ligado:
      print('Ligue o carro primeiro')
      return
    self.seta = direcao
    print(f'Seta lifada para {self.seta}')

#-----------------------------------------------
#metodos
carro1 = Carro('Preto', 2021)
carro1.informacoes()
carro1.ligar()
carro1.ligar_seta('esquerda')

```
