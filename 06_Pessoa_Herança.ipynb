```python
class Pessoa:
  def __init__(self, nome, idade):
    self.nome = nome
    self.idade = idade

  def apresentar(self):
    print(f'Meu nome é {self.nome} e minha idade é {self.idade}')

class Funcionario(Pessoa):
  def __init__(self, nome, idade, cargo):
    super().__init__(nome, idade)
    self.cargo = cargo

  def trabalhar(self):
    print(f'{self.nome} está trabalhando no cargo {self.cargo}')

class Cliente(Pessoa):
  def __init__(self, nome, idade, saldo):
    super().__init__(nome, idade)
    self.saldo = saldo

  def comprar(self, valor_compra):
    if valor_compra <= self.saldo:
      self.saldo -= valor_compra
      print(f'Olá {self.nome} sua compra de {valor_compra} foi aprovada! seu novo saldo é de {self.saldo}')
    else:
      print(f'A sua compra de {valor_compra} não foi aprovada! Saldo insuficiente!')
"""
f1 = Funcionario('João', 30, 'Gerente de Contas')
f1.apresentar()
f1.trabalhar()
"""
c1 = Cliente('Maria', 25, 200)
c2 = Cliente('José', 28, 100)
#c1.apresentar()
c1.comprar(80)
c2.comprar(20)
```
